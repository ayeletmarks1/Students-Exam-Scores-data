### 1. Arranging data
[back to top](#top)

```python
url= 'https://raw.githubusercontent.com/ayeletmarks1/Students-Exam-Scores-data/main/Expanded_data_with_more_features.csv'
student_df= pd.read_csv(url)
student_df.head()
```

<a id='section1.1'></a>
#### 1.1 Fix columns

```python
student_df=student_df.rename(columns={'Unnamed: 0': 'StudentNumber'})
student_df.head()
```

```python
student_df.shape
```

<a id='section1.2'></a>
#### 1.2 Creating column
[back to top](#top)

.הוספנו עמודה חדשה בה חישבנו את ממוצע הציונים במקצועות המתמטיקה, הקריאה והכתיבה עבור כל סטודנט

```python
student_df["avg_score"]= (student_df['MathScore'] + student_df['ReadingScore']+student_df['WritingScore'])/3
student_df.head(5)
```

<a id='section1.3'></a>
#### 1.3 Missing values
[back to top](#top)

```python
student_df= student_df.fillna("Unknown")
student_df.head(5)
```

```python
features_melt=['StudentNumber','Gender','TestPrep','PracticeSport','TransportMeans','WklyStudyHours']

melt_data= pd.melt(student_df[features_melt], id_vars = 'StudentNumber', value_vars = features_melt)

columns = sns.FacetGrid(data = melt_data[melt_data != 'Unknown'], col="variable",  col_wrap=5, sharex = False, sharey= False, height = 3)
columns.map_dataframe(sns.histplot, x = 'value', color="pink")
columns.set_xticklabels(rotation=15)
plt.subplots_adjust(wspace=0.8)
plt.show()
```

```python
features_melt2=['StudentNumber','MathScore','ReadingScore','WritingScore','avg_score']

melt_data= pd.melt(student_df[features_melt2], id_vars = 'StudentNumber', value_vars = features_melt2)

grades = sns.FacetGrid(data = melt_data[melt_data != 'Unknown'], col="variable",  col_wrap=5, sharex = False, sharey= False, height = 3)
grades.map_dataframe(sns.histplot, x = 'value', color="turquoise")
plt.subplots_adjust(wspace=0.8)
grades.set(xlim=(0, 100))
plt.show()
```

#### 2.1 First conclution
[back to top](#top)

```python
marital_status= sns.FacetGrid(data = student_df[student_df.ParentMaritalStatus != 'Unknown'], col="ParentMaritalStatus", hue='ParentMaritalStatus', col_wrap=4)
marital_status.map_dataframe(sns.histplot, x = 'avg_score', stat='percent')
plt.show()
```

בשלב זה נראה את הפיזור של כל הדאטה באחוזים לפי המצב המשפחתי של הסטודנט. למרות שניתן לראות שמעט מהסטודנטים קיבלנו ציון נמוך מ- 40, בחרנו להראות את השינוי ע"י בוקספלוט בצורה יותר ויזואלית 

```python
fig = plt.subplots(figsize=(5,6))
sns.boxplot(x='ParentMaritalStatus', y='avg_score', data=student_df[student_df.ParentMaritalStatus != 'Unknown'], palette= "gist_ncar_r")
plt.show()
```

אף על פי שציפינו לראות שינוי מסויים בממוצעי הציונים בקרב תלמידים עם סטטוס משפחתי שונה, ניתן לראות שאין שינוי ניכר בחציון. לכן, החלטנו להתמקד בהשפעה על התלמידים החריגים (קבענו מהו ציון חריג על פי האיור- ממוצע ציונים הנמוך מ- 30)

```python
exceptions_df = student_df.copy()
exceptions_df= exceptions_df[exceptions_df.avg_score <= 30]
```

```python
exceptions_df.shape
```

ניתן לראות שבהתחלה היו 30 אלף שורות, ולאחר צמצום הדאטה והתמקדות בחריגים, כמות השורות ירדה ל- 158 

```python
parent_edu = sns.FacetGrid(data = student_df[student_df.ParentEduc!= 'Unknown'], col='ParentEduc', hue='ParentEduc', col_wrap=3)
parent_edu.map_dataframe(sns.histplot, x = 'avg_score', stat='percent')
parent_edu.set(xlim=(0, 100))
plt.show()
```

כעת נציג את כמות הסטודנטים עם הציונים החריגים לפי השכלת ההורים ונבדוק כיצד השכלת ההורים משפיעה על ממוצע הציונים   

```python
exceptions_df_filtered = exceptions_df[exceptions_df['ParentEduc'] != 'Unknown']
sorted_categories = exceptions_df_filtered['ParentEduc'].value_counts().sort_values(ascending=False).index

fig = plt.subplots(figsize=(8, 5))
sns.countplot(data=exceptions_df_filtered, x='ParentEduc', palette='Reds', order=sorted_categories)
plt.xticks(rotation=50)
plt.show()
```

 כמות האחים במשפחה משפיעה על הציון כאשר אנו דנות בשאלה האם התלמיד הוא אח בכור או לא. ניתן לראות שבטווח ה2-3 אחים אין פער גדול בין ממוצע האח הבכור לממוצע שאר האחים אך כאשר מספר האחים במשפחה גדל הציון הממוצע של האח הבכור (הגרף הכחול) גבוה יותר מהציון הממוצע של שאר האחים (הגרף הכתום).  

<a id='section2.3'></a>
#### 2.3 Third conclution
[back to top](#top)

בקובץ הנתונים שלנו מוצגים ציוני התלמידים בשלושה מקצועות- מתמטיקה, קריאה וכתיבה.
רצינו לראות האם קיימים הבדלים במדד ההצלחה (בטווח הציונים- 70-100) בין גברים ונשים.

```python
male_math_scores = student_df[student_df['Gender'] == 'male']['MathScore'].tolist()
female_math_scores = student_df[student_df['Gender'] == 'female']['MathScore'].tolist()
male_reading_scores= student_df[student_df['Gender'] == 'male']['ReadingScore'].tolist()
female_reading_scores= student_df[student_df['Gender'] == 'female']['ReadingScore'].tolist()
male_writing_scores= student_df[student_df['Gender'] == 'male']['WritingScore'].tolist()
female_writing_scores= student_df[student_df['Gender'] == 'female']['WritingScore'].tolist()
```

```python
fig, ax = plt.subplots(1, 3, figsize=(20, 10))
plt.subplots_adjust(wspace=0.5)
ax[0].hist([male_math_scores, female_math_scores], bins=[70 ,80, 90, 100], rwidth=0.8, label=['male', 'female'])
ax[0].set_xlabel('Math Scores')
ax[0].set_ylabel('Number of Students')
ax[0].set_title('Distribution of Math scores by gender')
ax[0].legend()
ax[0].set_ylim([0, 4000])  
ax[0].set_xlim([70, 100])

ax[1].hist([male_reading_scores, female_reading_scores], bins=[ 70 ,80, 90, 100], rwidth=0.8, label=['male', 'female'])
ax[1].set_xlabel('Reading Scores')
ax[1].set_ylabel('Number of Students')
ax[1].set_title('Distribution of Reading scores by gender')
ax[1].legend()
ax[1].set_ylim([0, 4000])
ax[1].set_xlim([70, 100])

ax[2].hist([male_writing_scores, female_writing_scores], bins=[70 ,80, 90, 100], rwidth=0.8, label=['male', 'female'])
ax[2].set_xlabel('Writing Scores')
ax[2].set_ylabel('Number of Students')
ax[2].set_title('Distribution of Writing scores by gender')
ax[2].legend()
ax[2].set_ylim([0, 4000])
ax[2].set_xlim([70, 100])
plt.show()
```
